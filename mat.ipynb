```python
from matplotlib import pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[40,30,20,54]
plt.xlabel("languages",fontsize=15)
plt.ylabel("NO")
plt.title("Popularity")
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color="r",align="center",linewidth=4,alpha=0.7
        ,label="popularity")
plt.bar(x,z,width=0.4,color="y",align="center",linewidth=4,alpha=0.7
        ,label="popularity1")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np 

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[40,30,20,54]

width=8
p=np.arange(len(x))  
p1=[j+width for j in p  ]

plt.xlabel("languages",fontsize=15)
plt.ylabel("NO")
plt.title("Popularity")

plt.scatter(p,y,width,color="r"
        ,label="popularity")
plt.scatter(p1,z,width,color="y"
        ,label="popularity1")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```
